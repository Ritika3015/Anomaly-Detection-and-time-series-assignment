1) What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

- Anomaly Detection (also known as outlier detection) is the process of identifying data points, events, or observations that deviate significantly from the majority of the data. These anomalies could indicate critical incidents, such as:

- Fraud detection in banking
- Fault detection in systems
- Intrusion detection in cybersecurity
- Health monitoring in medical diagnostics

Types of Anomalies:

1. Point Anomaly (Global Outlier)

Definition: A single data point that is significantly different from the rest of the data.

Example:
In a list of daily credit card transactions:
[20, 25, 18, 22, 19, 5000, 24]
→ 5000 is a point anomaly — it deviates sharply from the normal spending pattern.

Use Case: Credit card fraud, temperature spikes, faulty sensor reading

2. Contextual Anomaly (Conditional Anomaly)

Definition: A data point is anomalous in a specific context (e.g., time, location, or environment), even though it might not be an outlier globally.

Example:
A very high CPU usage at midnight is abnormal, but during software updates at noon, it’s normal.

Use Case:

- Time-series analysis (e.g., seasonal sales trends)
- Weather monitoring

Energy consumption tracking

3. Collective Anomaly (Group Anomaly)

Definition: A group (collection) of related data points is anomalous together, even though the individual points may not be.

Example:
- A sudden burst of network traffic over a 5-minute window could indicate a DDoS attack, even though no single data point is extreme.
- A series of small, irregular heartbeat intervals may indicate an arrhythmia.

Use Case:
- Cybersecurity (network intrusions)
- Health monitoring (ECG patterns)
- Industrial process faults

2) Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.

1. Isolation Forest (iForest)
Approach:
- Based on the idea that anomalies are easier to isolate than normal points.
- Uses randomly constructed decision trees to partition the data.
- The fewer splits needed to isolate a point, the more likely it is an anomaly.

Characteristics:
- Feature	Description
- Type	Ensemble method (tree-based)
- Assumes Data Shape	No assumptions about distribution
- Handles High-Dimensional Data	Yes, effectively
- Output	Anomaly scores per data point
Efficiency	Very fast and scalable

Suitable Use Cases:
- High-dimensional datasets (e.g., fraud detection, cybersecurity)
- Large-scale anomaly detection
- Real-time anomaly detection systems

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
Approach:

- Groups data points into dense clusters.
- Points not belonging to any cluster (low density regions) are considered anomalies.
- Requires two parameters: eps (neighborhood radius) and minPts (minimum points to form a cluster)

Characteristics:
- Feature	Description
Type	Density-based clustering algorithm
- Assumes Data Shape	No specific shape, flexible clusters
- Handles High-Dimensional Data	Not ideal (suffers from curse of dimensionality)
- Output	Clusters + noise points (outliers)
- Efficiency	Moderate (depends on neighborhood search)

Suitable Use Cases:
- Spatial anomaly detection (e.g., GPS, geospatial data)
- Low-dimensional data with clusters and noise
- Applications where anomalies are isolated from dense regions

3. Local Outlier Factor (LOF)
Approach:
- Measures the local density of data points.
- A point is considered an anomaly if its local density is significantly lower than its neighbors.
- Compares how isolated a point is relative to its neighbors.

Characteristics:
- Feature	Description
- Type	Density-based anomaly detection
- Assumes Data Shape	Works best in clustered data
- Handles High-Dimensional Data	Not very well
- Output	LOF scores (higher = more anomalous)
- Efficiency	Slower on large datasets

Suitable Use Cases:
- Local anomalies in clustered datasets
- Detecting anomalies in time-series or transactional data
- Scenarios where the notion of "local" deviation is critical

3) What are the key components of a Time Series? Explain each with one
example.
A time series is a sequence of data points collected or recorded at regular time intervals (e.g., hourly, daily, monthly). Understanding the key components of a time series helps in analyzing patterns and building forecasting models.

There are four main components:

1. Trend (T)- The long-term direction in which the data is moving—upward, downward, or flat—over time.

Example:
- The average temperature rising steadily over 50 years due to climate change.
- A company’s sales revenue increasing gradually year over year.

2. Seasonality (S)- A repeating pattern or fluctuation in data at regular intervals (e.g., daily, weekly, monthly).

Example:
- Ice cream sales peak every summer and drop in winter.
- Electricity usage increasing every evening when people return home.
- Seasonality = regular, predictable cycles (typically calendar-based).

3. Cyclic Component (C)- Long-term fluctuations in a time series not of fixed frequency, often tied to economic or business cycles.

Example:
- Real estate prices rising and falling due to economic booms and recessions.
- Stock market trends influenced by political or macroeconomic events.
- Cycle = irregular, longer-term waves (not fixed like seasons).

4. Irregular or Residual/Noise (I or R)

Definition: Random, unpredictable variations in the data that cannot be attributed to trend, seasonality, or cycle.

Example:
- Sudden drop in website traffic due to a server outage.
- Unexpected sales spike after a celebrity mentions a product.
- Irregular = random “noise” or anomalies.

4) Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

A stationary time series is one whose statistical properties — like mean, variance, and autocorrelation — do not change over time.

To Check for Stationarity:
1. Visual Inspection
- Plot the data: Look for obvious trends or changing variance.
- If the series increases/decreases over time → likely non-stationary.
- Rolling mean/variance plots can help visualize changes.

2. Statistical Tests
Test	Description
- ADF Test (Augmented Dickey-Fuller)	Tests if a unit root is present (non-stationary if p > 0.05)
- KPSS Test (Kwiatkowski-Phillips-Schmidt-Shin)	Tests if the series is stationary around a trend (opposite of ADF)
- PP Test (Phillips-Perron)	Similar to ADF, tests for unit root

Rule of Thumb:
- ADF test: p-value < 0.05 → reject null → likely stationary
- KPSS test: p-value > 0.05 → fail to reject null → likely stationary

How to Transform a Non-Stationary Series into a Stationary One
1. Differencing- Subtract the previous value from the current value. This removes trend or seasonality. First-order differencing is usually enough.

2. Log Transformation- Helps stabilize variance.

3. Detrending- Subtract a trend line (linear regression) or moving average.

4. Seasonal Differencing- Subtract the value from the same season in the previous cycle.

5. Box-Cox Transformation- A power transformation to stabilize variance and normalize the data.

5) Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.

1. AR (AutoRegressive) Model
Structure:
- Predicts the current value based on past values.

Application:
- Used when past values are highly correlated with future values.
- Best for stationary time series without noise or seasonal patterns.

2. MA (Moving Average) Model
Structure:
- Predicts the current value based on past forecast errors.

Application:
- Used when the time series has irregular fluctuations or noise.
- Helps smooth short-term fluctuations.

3. ARIMA (AutoRegressive Integrated Moving Average)
Structure:
- Combines AR, MA, and differencing (I) to make a non-stationary series stationary.

Application:
- Suitable for non-seasonal, non-stationary time series with trends.
- General-purpose forecasting model in business, economics, etc.

4. SARIMA (Seasonal ARIMA)
Structure:
- Extends ARIMA by adding seasonal components.

Application:
- Best for data with seasonal trends (e.g., retail sales, temperature, web traffic).
- Can model both short-term and seasonal patterns.

5. SARIMAX (Seasonal ARIMA with eXogenous variables)
Structure:
- SARIMA model with external (exogenous) variables.

Application:
- When time series is influenced by external factors.
- E.g., sales influenced by ad spend, weather, holidays, etc.
- More flexible and realistic for real-world forecasting.



```python
#6) Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Plot original time series
plt.figure(figsize=(10,4))
plt.plot(data, label="AirPassengers")
plt.title("Monthly Number of Air Passengers")
plt.xlabel("Date")
plt.ylabel("Passengers")
plt.legend()
plt.show()

# Decompose time series
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative')

# Plot decomposition
decomposition.plot()
plt.show()
```

```python
#7) Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Generate synthetic taxi-like data
np.random.seed(42)
trip_distance = np.random.normal(5, 2, 500)   # average trip ~ 5 miles
fare_amount = trip_distance * 3 + np.random.normal(0, 5, 500)  # fare ~ distance * 3 + noise

# Add some anomalies (very high fares or mismatched values)
trip_distance = np.append(trip_distance, [2, 15, 20, 1, 30])
fare_amount = np.append(fare_amount, [200, 5, 400, 150, 600])

df = pd.DataFrame({"trip_distance": trip_distance, "fare_amount": fare_amount})

# Isolation Forest
iso = IsolationForest(contamination=0.05, random_state=42)
df['anomaly'] = iso.fit_predict(df[['trip_distance', 'fare_amount']])
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})

# Plot
plt.figure(figsize=(8,6))
plt.scatter(df['trip_distance'], df['fare_amount'],
            c=df['anomaly'], cmap="coolwarm", s=40, edgecolors="k")
plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.title("Isolation Forest Anomaly Detection (Synthetic Taxi Data)")
plt.colorbar(label="Anomaly (1 = Yes, 0 = No)")
plt.show()
```

```python
#8) Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Define SARIMA model (p,d,q)(P,D,Q,s)
# Here, s=12 for monthly seasonality
model = SARIMAX(data['Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_ci = forecast.conf_int()

# Plot results
plt.figure(figsize=(10,6))
plt.plot(data.index, data['Passengers'], label="Observed")
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, label="Forecast", color='red')
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color='pink', alpha=0.3)
plt.title("SARIMA Forecast of Airline Passengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

```python
#9) Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Generate synthetic 2D dataset
np.random.seed(42)
X_inliers = 0.5 * np.random.randn(200, 2)
X_inliers = np.r_[X_inliers + 2, X_inliers - 2]  # two clusters

# Add anomalies
X_outliers = np.random.uniform(low=-6, high=6, size=(20, 2))

# Combine data
X = np.r_[X_inliers, X_outliers]

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)
anomalies = y_pred == -1

# Plot results
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c='b', s=20, label="Normal")
plt.scatter(X[anomalies, 0], X[anomalies, 1], c='r', s=50, edgecolors='k', label="Anomalies")
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

```

10)  You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:
● How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?
● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?
● How would you validate and monitor the performance over time?
● How would this solution help business decisions or operations?

1. Anomaly Detection in Streaming Data
Goal: Detect sudden spikes/drops in energy usage every 15 minutes.
Approach: Use unsupervised anomaly detection methods since labeled anomalies are rare.
- Isolation Forest (IF): Good for high-dimensional data and handles non-linear patterns; can quickly flag unusual consumption points.
- Local Outlier Factor (LOF): Captures local density deviations; useful if anomalies are relative to neighbors (e.g., one region behaving differently).
- DBSCAN: Good for clustering + anomaly detection; points that don’t belong to clusters are anomalies. Works well if consumption patterns cluster naturally.

Additional Safeguards:
- Use rolling z-scores or percentile thresholds to catch sudden spikes beyond historical seasonal patterns.
- Combine model-based detection (IF/LOF) with rule-based alerts (e.g., >3σ deviation, >30% sudden drop).

2. Time Series Forecasting for Short-Term Demand
Forecast horizon: Next few hours to next day (important for load balancing).
Models:
- ARIMA: Handles short-term trends but not seasonality.
- SARIMA: Better if daily/weekly seasonal cycles exist.
- SARIMAX: Even better, since it allows exogenous features (weather, region). Energy usage is strongly dependent on temperature, humidity, day of week, holidays → SARIMAX is the most suitable.

Alternative:
- Facebook Prophet or LSTM/RNNs can be considered if data is very large and seasonality is complex.

3. Validation and Performance Monitoring

Validation approach:
- Use time series cross-validation (rolling/expanding window) rather than random splits.
- Evaluate using MAPE, RMSE, MAE (MAPE is business-friendly since it’s % error).
- Monitoring in production: Implement drift detection: Monitor if forecast errors are systematically increasing (concept drift due to new usage patterns).
- Use real-time dashboards (e.g., Power BI, Grafana, Kibana) to compare predicted vs actual.
- Trigger model retraining periodically (weekly/monthly) or when error exceeds a threshold.

4. Business & Operational Impact

- Grid Reliability: Early anomaly detection → prevent blackouts by identifying load spikes before equipment stress.
- Cost Savings: Improved forecasting → reduce reliance on expensive reserve energy sources.
- Operational Efficiency: Load balancing across regions → optimize generation and distribution.
- Customer Experience: Early detection of abnormal usage patterns → flag potential faulty meters, theft, or unusual demand surges.
- Strategic Planning: Seasonal demand forecasting → helps with long-term planning for capacity and infrastructure.
